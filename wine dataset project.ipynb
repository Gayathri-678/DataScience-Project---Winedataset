```python
#importing data visualisation libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(x='alcohol',y='quality',data=fb)
```

```python
plt.scatter(fb['alcohol'],fb['quality'])
plt.xlabel('alcohol')
plt.ylabel('quality')
plt.show()
```

```python

count,bin_edges=np.histogram(fb["alcohol"])
fb["alcohol"].plot(kind="hist",xticks=bin_edges)
plt.xlabel("value of alcohal")
plt.ylabel("quality")
plt.grid()
plt.show()
```

```python
auto_temp=fb[["chlorides","sulphates","quality"]]
auto_group=auto_temp.groupby(["chlorides","sulphates"],as_index=False).mean()
auto_group
```
